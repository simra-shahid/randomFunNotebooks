```python
import pandas as pd
import numpy as np

movies=pd.read_csv("fandango_score_comparison.csv")
movies
```

```python
%matplotlib inline 
import matplotlib.pyplot as plt


plt.hist(movies["Metacritic_norm_round"], alpha=0.5,color="orange")
plt.title("Metacritic")
plt.xlim(0,5)
plt.show()

plt.hist(movies["Fandango_Stars"],alpha=0.5, color="yellow")
plt.title("Fandango_Stars")
plt.xlim(0,5)
plt.show()


plt.hist(movies["IMDB_norm"], alpha=0.5,color="orange")
plt.title("IMDB")
plt.xlim(0,5)
plt.show()

plt.hist(movies["RT_user_norm"],alpha=0.5, color="yellow")
plt.title("Rotten Tomatoes")
plt.xlim(0,5)
plt.show()
```

In Fandago there are no ratings less than 3 and most of them are centered around 4.0 and 4.5 
In Metacritic most of the ratings are around 3.5-4.5
In IMBD there are no ratings less than 2. Most of the ratings are centered between 3-4
In Rotten Tomatoes most of the ratings are centered around 4


```python
f_mean=movies["Fandango_Stars"].mean()
m_mean=movies["Metacritic_norm_round"].mean()
f_median=movies["Fandango_Stars"].median()
m_median=movies["Metacritic_norm_round"].median()
f_std=movies["Fandango_Stars"].std()
m_std=movies["Metacritic_norm_round"].std()

print("Fandango Mean   =%f       Metacritic Mean   =%f"%(f_mean,m_mean))
print("Fandango Median =%f       Metacritic Median =%f"%(f_median,m_median))
print("Fandango SD     =%f       Metacritic SD     =%f"%(f_std,m_std))
print()
print("IMBD Mean       =%f       Rotten Tomatoes Mean   =%f"%(movies["IMDB_norm"].mean(),movies["RT_user_norm"].mean()))
print("IMBD Median     =%f       Rotten Tomatoes Median =%f"%(movies["IMDB_norm"].median(),movies["RT_user_norm"].median()))
print("IMBD SD         =%f       Rotten Tomatoes SD     =%f"%(movies["IMDB_norm"].std(),movies["RT_user_norm"].std()))

```

```python
##Fandango vs Metacritic Methodology
Fandango appears to inflate ratings and isn't transparent about how it calculates and aggregates ratings.
Metacritic publishes each individual critic rating, and is transparent about how they aggregate them to get a final rating.
Like Metacrtic, IMBD and Rotten Tomatoes are transparent.
```

```python
##Fandango vs Metacritic number differences
The median metacritic score appears higher than the mean metacritic score because a few very low reviews "drag down" the median.
The median fandango score is lower than the mean fandango score because a few very high ratings "drag up" the mean.

Fandango ratings appear clustered between 3 and 5, and have a much narrower random than Metacritic reviews, which go from 0 to 5.
Fandango ratings in general appear to be higher than metacritic ratings.
These may be due to movie studio influence on Fandango ratings, and the fact that Fandango calculates its ratings in a hidden way.
```

```python
from scipy.stats import pearsonr ,linregress
r_value,p_value=pearsonr(movies["Fandango_Stars"],movies["Metacritic_norm_round"])
r_value
```

```python
Low r_value for Fandango and Metacritic
```

```python
r_value,p_value=pearsonr(movies["IMDB_norm"],movies["RT_user_norm"])
r_value
```

```python
High r_value for IMBD and Rotten Tomatoes
```

```python
slope,intercept,r_value,p_value,stderr_slope=linregress(movies["Metacritic_norm_round"],movies["Fandango_Stars"])
#y=mx+c
pred_3 = 3 * slope + intercept
pred_3
```

```python
pred_1= 1 * slope + intercept
pred_1
```

```python
pred_5= 5 * slope + intercept
pred_5
```

```python
x_values=[1.0,5.0]
y_values=[pred_1,pred_5]
plt.plot(x_values,y_values,linewidth=2,color="green",alpha=0.5)
plt.scatter(movies["Metacritic_norm_round"],movies["Fandango_Stars"],color="orange")
plt.xlim(1,5)
plt.show()
```

```python
slope,intercept,r_value,p_value,stderr_slope=linregress(movies["IMDB_norm"],movies["RT_user_norm"])
#y=mx+c
pred_1 = 1 * slope + intercept
print("predicted value in Rotten Tomatoes with Linear Regression= ", pred_1)
pred_5 = 5 * slope + intercept
print("predicted value in Rotten Tomatoes with Linear Regression= ", pred_5)
```

```python
We can see that ratings in both the websites are close by
```
