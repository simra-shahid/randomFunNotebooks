```python
#GROUP BY
raw_data = {'first_name': ['Jason', 'Molly', 'Tina', 'Jake', 'Amy'],
        'pre_score': [4, 24, 31, 2, 3],
        'mid_score': [25, 94, 57, 62, 70],
        'post_score': [5, 43, 23, 23, 51]}
df = pd.DataFrame(raw_data, columns = ['first_name', 'pre_score', 'mid_score', 'post_score'])
df
```

```python

import numpy as np
pos=np.arange(len(df['pre_score']))
list(pos)
```

```python
pos=list(range(len(df['pre_score'])))
pos
```

```python
fig,ax=plt.subplots(figsize=(10,5))

plt.bar(pos,df['pre_score'],0.25,color="#EE3224",alpha=0.5)
pos1=[p+0.25 for p in pos]
plt.bar(pos1,df['mid_score'],0.25,color='#F78F1E',alpha=0.5)
pos2=[p+0.25 for p in pos1]
plt.bar(pos2,df['post_score'],0.25,color='#FFC222',alpha=0.5)

tick_positions=[p + 0.375 for p in pos]

ax.set_xticks(tick_positions)

ax.set_xticklabels(df['first_name'])

plt.xlim(min(pos)-0.25, max(pos)+0.25*4)

max_of_three=max(max(df['pre_score']),max(df['mid_score']),max(df['post_score']))+10

plt.ylim([0, max_of_three ] )

plt.legend(['Pre Score', 'Mid Score', 'Post Score'], loc='upper left')
plt.grid()
plt.show()
```

```python
# Grouped bar plot to compare the number of men to women in each category


fig,ax=plt.subplots(figsize=(100,100))

pos= list(range(20))
plt.bar(pos ,recent_grads['Men'][:20],0.25,color="#EE3224",alpha=0.5)

pos1=[p+0.25 for p in pos]
plt.bar(pos1,recent_grads['Women'][:20],0.25,color='#F78F1E',alpha=0.5)



tick_positions=[p + 0.25 for p in pos]

ax.set_xticks(tick_positions)

ax.set_xticklabels(recent_grads['Major'],rotation=90)

plt.xlim(min(pos)-0.25, max(pos)+0.25*4)

max_of_two=max(max(recent_grads['Men'][:20]),max(recent_grads['Women'][:20]))+1

plt.ylim([0, max_of_two ])

plt.legend(['Men', 'Women'], loc='upper left')
plt.grid()
plt.show()
```
