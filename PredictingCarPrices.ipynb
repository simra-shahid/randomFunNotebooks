```python
import pandas as pd
import numpy as np

pd.options.display.max_columns = 99
np.random.seed(1)
```

```python
cols=['symboling', 'normalized-losses', 'make', 'fuel-type', 'aspiration', 'num-of-doors', 'body-style', 
        'drive-wheels', 'engine-location', 'wheel-base', 'length', 'width', 'height', 'curb-weight', 'engine-type', 
        'num-of-cylinders', 'engine-size', 'fuel-system', 'bore', 'stroke', 'compression-rate', 'horsepower',
      'peak-rpm', 'city-mpg', 'highway-mpg', 'price']
cars=pd.read_csv("imports-85.data",names=cols)
cars.head()
```

```python
cars.replace("?",np.nan)
```

```python
continuous_values_cols = ['normalized-losses', 'wheel-base', 'length', 'width', 'height', 'curb-weight', 'bore', 'stroke', 'compression-rate', 'horsepower', 'peak-rpm', 'city-mpg', 'highway-mpg', 'price']
numeric_cars = cars[continuous_values_cols]
```

```python
numeric_cars=numeric_cars.replace("?",np.nan)
```

```python
numeric_cars.head()
```

```python
numeric_cars=numeric_cars.astype(float)
```

```python
numeric_cars.isnull().sum()
```

```python
# We will be predicting prices so we remove rows with missing prices
numeric_cars=numeric_cars.dropna(subset=["price"])
```

```python
numeric_cars.isnull().sum()
```

```python
#fill missings values in other columns by mean of that column

numeric_cars=numeric_cars.fillna(numeric_cars.mean())
```

```python
numeric_cars.isnull().sum()
```

```python
#normalize all columns  so all values range from 0 to 1 except target column(price)
price_col=numeric_cars["price"]
numeric_cars=(numeric_cars-numeric_cars.min())/(numeric_cars.max()-numeric_cars.min())
numeric_cars["price"]=price_col
numeric_cars.head()
```

# Train/test Model for k=5 for every column

```python
from sklearn.neighbors import KNeighborsRegressor
from sklearn.metrics import mean_squared_error
import random

result={}

def knn_train_test(features,target,df):
   
    np.random.seed(1)
    
    # Randomize order of rows in data frame.
    shuffled_index = np.random.permutation(df.index)
    rand_df = df.reindex(shuffled_index)

    # Divide number of rows in half and round.
    last_train_row = int(len(rand_df) / 2)
    
    # Select the first half and set as training set.
    # Select the second half and set as test set.
    train= rand_df.iloc[0:last_train_row]
    test= rand_df.iloc[last_train_row:]
    
    knn=KNeighborsRegressor()
    knn.fit(train[[features]],train[target])
    predictions=knn.predict(test[[features]])
    mse= mean_squared_error(test[target],predictions)
    rmse=mse**0.5
    return rmse

t=numeric_cars.columns.drop("price")
for cols in t:
    rmse=knn_train_test(cols,"price",numeric_cars)
    result[cols]=rmse
        
result_series=pd.Series(result)
result_series.sort_values()
        
        
        
```

# For all columns using k=1,3,5,7,9

```python
k_values=[1,3,5,7,9]


result={}

def knn_train_test(features,target,df):
     
        rmse_result={}
        np.random.seed(1)
    
        # Randomize order of rows in data frame.
        shuffled_index = np.random.permutation(df.index)
        rand_df = df.reindex(shuffled_index)

        # Divide number of rows in half and round.
        last_train_row = int(len(rand_df) / 2)

        # Select the first half and set as training set.
        # Select the second half and set as test set.
        train= rand_df.iloc[0:last_train_row]
        test= rand_df.iloc[last_train_row:]

        for k in k_values:
            knn=KNeighborsRegressor(n_neighbors=k)
            knn.fit(train[[features]],train[target])
            predictions=knn.predict(test[[features]])
            mse= mean_squared_error(test[target],predictions)
            rmse=mse**0.5
            rmse_result[k]=rmse
        return rmse_result

t=numeric_cars.columns.drop("price")

for cols in t:
    rmse=knn_train_test(cols,"price",numeric_cars)
    result[cols]=rmse

result 
        
```

# Using k-fold cross validation 

#Univariate for k=5

```python
from sklearn.model_selection import KFold,cross_val_score

result={}

def knn_kfold(features,target,df):
     
        kf=KFold(5,shuffle=True, random_state=1)
        knn=KNeighborsRegressor()
        mses=cross_val_score(knn,df[[features]],df[[target]],
                             scoring="neg_mean_squared_error",cv=kf)
        rmse=np.sqrt(np.absolute(mses))
        avg_rmse=np.mean(rmse)
        return avg_rmse

t=numeric_cars.columns.drop("price")

for cols in t:
    rmse=knn_kfold(cols,"price",numeric_cars)
    result[cols]=rmse

result_series=pd.Series(result)
result_series.sort_values()
```
